### B.  Describe multiple regression methods by doing the following:

<b>1.  Summarize the assumptions of a multiple regression model.</b>

When building a multiple regression model there are a few assumptions you must consider. First, there should be a linear relationship between the response variable and each of the predictor variables. The second assumption is that there is no multicollinearity betweent the predictor variables. This means if you plot a scatter plot between any two predictor variables the points should appear mostly random. Third, the residuals created from the model should fall into a normal distribution. Finally, the model's residuals should appear homoscedastic. Creating a plot of the residuals should, ideally, show a seemingly random distribution across the range of the of all the residuals.

```python
# Import necessary packages
# Import necessary packages
import numpy as np
import pandas as pd
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor 
from sklearn import metrics
from sklearn import linear_model
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from scipy.stats import pearsonr

# Import Visualization Tools
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# Read dataset into a dataframe
df = pd.read_csv('churn_clean.csv')

# List and drop unneeded columns/variables
drop_list = ['Item1', 'Item2', 'Item3', 'Item4', 'Item5', 'Item6', 'Item7', 'Item8',
             'CaseOrder', 'Customer_id', 'Interaction', 'UID', 'City', 'State',
             'County', 'Zip', 'Lat', 'Lng', 'Population', 'Area', 'TimeZone', 'Job',
             'Churn','Income','OnlineBackup', 'DeviceProtection','Marital', 
             'PaymentMethod', 'Contract','Email','Outage_sec_perweek','PaperlessBilling']

df = df.drop(columns = drop_list)

# List and convert Categorical variables to dummy variables
to_dummy_list = ['Gender', 'InternetService']

df = pd.get_dummies(data=df, columns=to_dummy_list, drop_first=True)

# Convert "yes" and "no" to 1 and 0 respectively
df = df.replace(to_replace='Yes', value=1)
df = df.replace(to_replace='No',value=0)

df.columns
```

# Part IV: Model Comparison and Analysis

### D.  Compare an initial and a reduced multiple regression model by doing the following:

<b>1.  Construct an initial multiple regression model from all predictors that were identified in Part C2.</b>

```python
# Separate data into predictors (X) and response (y) 
X = df.drop(columns=['Bandwidth_GB_Year'], axis=1)
y = df['Bandwidth_GB_Year']

# Add column for constant to the X matrix
Xc = sm.add_constant(X)

# Fit model
lin_reg = sm.OLS(y,Xc)
lin_mod = lin_reg.fit()

# Show summary of model
lin_mod.summary()
```

<b>2.  Justify a statistically based variable selection procedure and a model evaluation metric to reduce the initial model in a way that aligns with the research question.</b>

```python
# Check Variance Inflation Factors to see what variable are related
# Code sourced from Stack Overflow https://stackoverflow.com/questions/42658379/variance-inflation-factor-in-python
pd.Series([variance_inflation_factor(Xc.values, i) for i in range(Xc.shape[1])], index=Xc.columns)
```

```python
Xs = StandardScaler().fit_transform(X)
pcamodel = PCA(n_components=11)
pca = pcamodel.fit_transform(Xs)

# Add column for constant to the X matrix
X_pca = sm.add_constant(pca)

# Fit model
lin_reg2 = sm.OLS(y,X_pca)
lin_mod2 = lin_reg2.fit()

# Show summary of model
lin_mod2.summary()
```

```python
x_ticks = np.arange(pcamodel.n_components_)+1
plt.plot(x_ticks,pcamodel.explained_variance_)
plt.title('Scree Plot')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance')
plt.show();
```

```python
# Check for Normal Distribution of Residuals
residuals = y - lin_mod2.predict(X_pca)
plt.title('Distribution of Residuals')
sns.distplot(residuals);
```

```python
# Check for Multicollinearity
pca_df = pd.DataFrame(pca)
sns.heatmap(pca_df.corr());
```

```python
# Test for Heterscedasticity
sns.scatterplot(x=pca_df.index, y=residuals,alpha=.25)
plt.title('Residuals');
```
